```python
# Getting unique values in the 'City' column
unique_cities = df['City'].unique()

# Display the unique values
print(unique_cities)
```

```python
# Mapping of city names to country names
city_to_country = {
    'Bengaluru': 'Germany',
    'Delhi': 'France',
    'Mumbai': 'Spain',
    'Pune': 'Italy',
    'Chennai': 'United Kingdom',
    'Kolkata': 'Netherlands',
    'Hyderabad': 'Sweden'
}

# Replacing city names with country names
df1['City'] = df1['City'].replace(city_to_country)

# Now, df has the updated city names as European countries
df1.sample(10)
```

```python
df1.rename(columns={'City': 'Country'}, inplace=True)

# Display the DataFrame to confirm the change
df1.sample(10)
```

```python
df1['Opportunity Status'] = df1['Opportunity Status'].replace('Loss', 'Lost')
df1.sample(10)
```
