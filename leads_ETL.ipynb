```python
import pandas as pd
import random

# Determine the total number of rows in the dataset
total_rows = len(df)

# Provided counts for "Technology Primary"
tech_primary_counts = {
    "Analytics": 37,
    "ERP Implementation": 7659,
    "Legacy Modernization": 101,
    "Technical Business Solutions": 4203
}

# Calculate total count for "Technology Primary"
total_tech_primary = sum(tech_primary_counts.values())

# Calculate the fraction of each "Technology Primary" category
tech_primary_fractions = {k: v / total_tech_primary for k, v in tech_primary_counts.items()}

# Create weighted choice list
weighted_tech_primary = []
for tech, fraction in tech_primary_fractions.items():
    weighted_tech_primary.extend([tech] * int(fraction * 10000))  # multiplying by 10000 for better granularity

# Randomly assign "Product Interested" values based on "Technology Primary" distribution
random.seed(42)  # for reproducibility
df['Product Interested'] = [random.choice(weighted_tech_primary) for _ in range(len(df))]

# If you want to save the updated dataset to a new CSV file
df.to_csv('Updated_lead_Dataset.csv', index=False)
```
